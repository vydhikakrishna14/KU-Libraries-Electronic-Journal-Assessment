```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('/content/Dataset-4.xlsx')

# Get the counts of each title
title_counts = df['subject'].value_counts()

# Calculate the percentage of each title
title_percentage = (title_counts / title_counts.sum()) * 100

# Plotting
plt.figure(figsize=(10, 6))
title_percentage.plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Percentage of Titles')
plt.xlabel('Titles')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('/content/Dataset-4.xlsx')

# Define the subject attributes
subject_attributes = ['HISTORY - HISTORY OF AFRICA',
                      "WOMEN'S HEALTH",
                      'HISTORY - HISTORY OF NORTH AND SOUTH AMERICA',
                      'NATIVE AMERICAN STUDIES',
                      'ASIAN STUDIES',
                      "WOMEN'S STUDIES",
                      'HISTORY - HISTORY OF THE NEAR EAST',
                      'LGBTQ+',
                      'POLITICAL SCIENCE - CIVIL RIGHTS',
                      'LAW - CIVIL LAW',
                      "MEN'S STUDIES"]

# Initialize a dictionary to store counts for each subject attribute
title_counts = {subject: len(df[df['subject'] == subject]) for subject in subject_attributes}

# Calculate the total count of titles
total_count = sum(title_counts.values())

# Calculate the percentage of titles for each subject attribute
title_percentage = {subject: (count / total_count) * 100 for subject, count in title_counts.items()}

# Sort the title_percentage dictionary by values in descending order
sorted_title_percentage = dict(sorted(title_percentage.items(), key=lambda item: item[1], reverse=True))

# Print the information
print("Percentage of Titles Each Subject Attribute Owns (in descending order):")
for subject, percentage in sorted_title_percentage.items():
    print(f"{subject}: {percentage:.2f}%")
```

***USING PUBLISHER ATTRIBUTE***

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/content/Dataset-4.xlsx')

# Get the counts of each publisher
publisher_counts = df['publisher'].value_counts()

# Calculate the percentage of each publisher
publisher_percentage = (publisher_counts / publisher_counts.sum()) * 100

# Plotting with exchanged axes
plt.figure(figsize=(10, 6))
publisher_percentage.plot(kind='barh', color='lightpink', edgecolor='purple')  # Using barh() for horizontal bar plot
plt.title('Percentage of Titles by Publisher')
plt.ylabel('Publisher')  # Switching x and y labels
plt.xlabel('Percentage')  # Switching x and y labels
plt.grid(axis='x')  # Switching grid axis
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('/content/Dataset-4.xlsx')

# Get the counts of each publisher
publisher_counts = df['publisher'].value_counts()

# Calculate the percentage of each publisher
publisher_percentage = (publisher_counts / publisher_counts.sum()) * 100

# Sort the publisher_percentage Series by values in descending order
sorted_publisher_percentage = publisher_percentage.sort_values(ascending=False)

# Print the information
print("Percentage of Titles Each Publisher Owns (in descending order):")
for publisher, percentage in sorted_publisher_percentage.items():
    print(f"{publisher}: {percentage:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/content/Dataset-4.xlsx')

# Get the counts of each publisher
publisher_counts = df['publisher'].value_counts()

# Calculate the percentage of each publisher
publisher_percentage = (publisher_counts / publisher_counts.sum()) * 100

# Plotting with exchanged axes and space between y-axis labels
plt.figure(figsize=(10, 6))
plt.plot(publisher_percentage.values, publisher_percentage.index, marker='o', color='skyblue', linestyle='-')  # Exchanging x and y values
plt.title('Percentage of Titles by Publisher')
plt.xlabel('Percentage')  # Switching x and y labels
plt.ylabel('Publisher')  # Switching x and y labels
plt.yticks(publisher_percentage.index, [f'{publisher}\n' for publisher in publisher_percentage.index])  # Adding space between titles on y-axis
plt.xticks(rotation=45, ha='right')
plt.grid(True)

# Annotate each point with its respective percentage value
for publisher, percentage in zip(publisher_percentage.index, publisher_percentage.values):
    plt.annotate(f'{percentage:.2f}%', xy=(percentage, publisher), xytext=(5, 5), textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Read the dataset
df = pd.read_excel('/content/Dataset-4.xlsx')

# Get the counts of each publisher
publisher_counts = df['publisher'].value_counts()

# Calculate the percentage of each publisher
publisher_percentage = (publisher_counts / publisher_counts.sum()) * 100

# Create a DataFrame to display the results
results_df = pd.DataFrame({
    'Publisher': publisher_percentage.index,
    'Percentage': publisher_percentage.values
})

# Sort the DataFrame by the percentage values in descending order
results_df = results_df.sort_values(by='Percentage', ascending=False)

# Display the results in a table format
print(results_df)
```
