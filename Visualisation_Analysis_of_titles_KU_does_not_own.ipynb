```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/5-1.csv')

# Define the subject attributes
subject_attributes = ['HISTORY - HISTORY OF AFRICA',
                      "WOMEN'S HEALTH",
                      'HISTORY - HISTORY OF NORTH AND SOUTH AMERICA',
                      'NATIVE AMERICAN STUDIES',
                      'ASIAN STUDIES',
                      "WOMEN'S STUDIES",
                      'HISTORY - HISTORY OF THE NEAR EAST',
                      'LGBTQ+',
                      'POLITICAL SCIENCE - CIVIL RIGHTS',
                      'LAW - CIVIL LAW',
                      "MEN'S STUDIES",
                      "WOMEN'S INTERESTS"]

# Initialize a dictionary to store counts for each subject attribute
title_counts = {subject: len(df[df['subject'] == subject]) for subject in subject_attributes}

# Calculate the total count of titles
total_count = sum(title_counts.values())

# Calculate the percentage of titles for each subject attribute
title_percentage = {subject: (count / total_count) * 100 for subject, count in title_counts.items()}

# Sort the title_percentage dictionary by values in descending order
sorted_title_percentage = dict(sorted(title_percentage.items(), key=lambda item: item[1], reverse=True))

# Print the information
print("Percentage of Titles Each Subject Attribute Owns (in descending order):")
for subject, percentage in sorted_title_percentage.items():
    print(f"{subject}: {percentage:.2f}%")
```

```python
import pandas as pd


df = pd.read_csv('/content/5-1.csv')

# Define the subject attributes
subject_attributes = ['HISTORY - HISTORY OF AFRICA',
                      "WOMEN'S HEALTH",
                      'HISTORY - HISTORY OF NORTH AND SOUTH AMERICA',
                      'NATIVE AMERICAN STUDIES',
                      'ASIAN STUDIES',
                      "WOMEN'S STUDIES",
                      'HISTORY - HISTORY OF THE NEAR EAST',
                      'LGBTQ+',
                      'POLITICAL SCIENCE - CIVIL RIGHTS',
                      'LAW - CIVIL LAW',
                      "MEN'S STUDIES",
                      "WOMEN'S INTEREST"]

# Initialize a dictionary to store counts for each subject attribute
title_counts = {subject: len(df[df['subject'] == subject]) for subject in subject_attributes}

# Calculate the total count of titles
total_count = sum(title_counts.values())

# Calculate the percentage of titles for each subject attribute
title_percentage = {subject: (count / total_count) * 100 for subject, count in title_counts.items()}

# Sort the title_percentage dictionary by values in descending order
sorted_title_percentage = dict(sorted(title_percentage.items(), key=lambda item: item[1], reverse=True))

# Check for overlapping percentages
duplicate_percentage = [percentage for percentage, count in title_percentage.items() if list(title_percentage.values()).count(percentage) > 1]

if duplicate_percentage:
    print("There are overlapping percentages.")
    print("Duplicate percentages:")
    print(duplicate_percentage)
    total_overlap_titles = sum(list(title_percentage.values()).count(percentage) for percentage in duplicate_percentage)
    print(f"Total number of titles which overlap: {total_overlap_titles}")
else:
    print("There are no overlapping percentages.")
```

***USING PUBLISHER ATTRIBUTE***

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace 'your_dataset.csv' with your actual dataset)
df = pd.read_csv('/content/5-1.csv')

# Assuming 'titles' is the column name containing titles
publishers = df['publisher']

# Get the counts of each publisher
publisher_counts = publishers.value_counts()

# Get unique publishers
unique_publishers = publisher_counts.index

# Get the frequency of each publisher
frequency = publisher_counts.values

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(unique_publishers, frequency, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Titles by Publisher')
plt.xlabel('Publisher')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
